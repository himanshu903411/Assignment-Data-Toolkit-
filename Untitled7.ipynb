```python
# 1.Demonstrate three different methods for creating identical 2D arrays in NumPy Provide the code for each method and the final output after each method?

#.1 Using np.array()

import numpy as np

# Method 1: Using np.array()
array_1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array 1:")
print(array_1)

# 2. 2. Using np.full()

# Method 2: Using np.full()
array_2 = np.full((2, 3), 0)  # Create a 2x3 array filled with zeros
array_2[0, :] = [1, 2, 3]     # Manually set values row by row
array_2[1, :] = [4, 5, 6]     # Manually set values row by row
print("\nArray 2:")
print(array_2)

# 3. Using np.zeros() and modifying elements.

# Method 3: Using np.zeros() and modifying elements
array_3 = np.zeros((2, 3), dtype=int)  # Create a 2x3 array filled with zeros
array_3[0] = [1, 2, 3]  # Assign values to the first row
array_3[1] = [4, 5, 6]  # Assign values to the second row
print("\nArray 3:")
print(array_3)

# 2. Using the Numpy function, generate an array of wRR evenly spaced numPers Petween w and wR and Reshape that wD array into a 2D array ?

import numpy as np

# Given values (you can modify these as needed)
w = 0        # Start value
wR = 10      # End value
numPoints = 12  # Number of points (evenly spaced)
rows = 4     # Number of rows for reshaping
cols = 3     # Number of columns for reshaping

# Step 1: Generate an array of numPoints evenly spaced between w and wR
array_1d = np.linspace(w, wR, numPoints)

# Step 2: Reshape the array into a 2D array (rows x cols)
array_2d = array_1d.reshape(rows, cols)

print("1D Array:")
print(array_1d)
print("\nReshaped 2D Array:")
print(array_2d)

# 3.Explain the following termsg
# .M The difference in npYarray, npYasarray and npYasanyarrayX
#. M The difference between Deep copy and shallow copyX

# 1. np.array()

import numpy as np

# Example
a = [1, 2, 3]
arr = np.array(a)

# 2. np.asarray()

# Example
arr = np.asarray([1, 2, 3])  # Converts to ndarray
existing_arr = np.asarray(arr)  # Does not create a new copy, just returns the same object

# 3. np.asanyarray()

# Example with matrix subclass
arr = np.matrix([[1, 2], [3, 4]])
asany_arr = np.asanyarray(arr)  # Retains the matrix subclass

# 4 . Generate a 3x3 array with random floating-point numPers Petween . and 5 and 20 then, round each numPer in  the array to 2 decimal places ?

import numpy as np

# Step 1: Generate a 3x3 array with random floats between 5 and 20
random_array = np.random.uniform(5, 20, (3, 3))

# Step 2: Round each number to 2 decimal places
rounded_array = np.round(random_array, 2)

print("Random Array:")
print(random_array)
print("\nRounded Array:")
print(rounded_array)

# 5. Create a NumPy array with random integers Petween 1 and 10 of shape (5, 6)) After creating the array perform the following operations:

 # a)Extract all even integers from array.

 # b)Extract all odd integers from array.

import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, size=(5, 6))
print("Original Array:\n", array)

# a) Extract all even integers from the array
even_integers = array[array % 2 == 0]
print("\nEven Integers:\n", even_integers)

# b) Extract all odd integers from the array
odd_integers = array[array % 2 != 0]
print("\nOdd Integers:\n", odd_integers)

# 6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers Petween 1 and 10 Perform the following operations:

 # a) Find the indices of the maximum values along each depth level (third axis).

# b) Perform element wise multiplication of between both arrayX import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) with random integers between 1 and 10
array_3d = np.random.randint(1, 11, size=(3, 3, 3))
print("3D Array:\n", array_3d)

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3d, axis=2)
print("\nIndices of Maximum Values along the Third Axis:\n", max_indices)

# b) Perform element-wise addition of a constant, e.g., adding 5 to each element
added_array = array_3d + 5
print("\nArray after Element-wise Addition of 5:\n", added_array)

# 7. Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type Also display the taPle attriPutes and data types of each column

import pandas as pd

# Sample dataset
data = {
    'Name': [' Shelia  ', ' jo ', 'Sheryl'],
    'Phone': ['857.139.8239 ', '0000000000', '(599)782-0605'],
    'Age': [25, 30, 35]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(df)

# Clean the 'Phone' column by removing non-numeric characters
df['Phone'] = df['Phone'].replace(r'\D', '', regex=True)  # Remove non-numeric characters

# Convert the 'Phone' column to a numeric data type
df['Phone'] = pd.to_numeric(df['Phone'], errors='coerce')  # Convert to numeric

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)

# Display the attributes and data types of each column
print("\nData Types of Each Column:")
print(df.dtypes)

# 10. Create a 7*9. Dataframe in Pandas using a series generated from 35. random integers Petween 1 to 6 )?

import pandas as pd
import numpy as np

# Generate 35 random integers between 1 and 6
random_numbers = np.random.randint(1, 7, size=35)

# Create a Series from the random numbers
random_series = pd.Series(random_numbers)

# Reshape the Series into a 7x9 DataFrame
# Since 35 elements cannot perfectly fit into a 7x9 DataFrame, we will create a 7x5 DataFrame and fill the rest with NaN
df = random_series.values.reshape(7, 5)  # Reshape it to 7 rows and 5 columns
df = pd.DataFrame(df)  # Convert to DataFrame

print(df)


#11. Create two different Series, each of length 50, with the following criteria:

#a) The first Series should contain random numbers ranging from 10 to 50.

#b) The second Series should contain random numbers ranging from 100 to 1000.

#c) Create a DataFrame by 'joining these Series by column, and, change the names of the columns to 'col1', 'col2',etc&

import pandas as pd
import numpy as np

# Step a: Create the first Series with random numbers ranging from 10 to 50
series1 = pd.Series(np.random.randint(10, 51, size=50))

# Step b: Create the second Series with random numbers ranging from 100 to 1000
series2 = pd.Series(np.random.randint(100, 1001, size=50))

# Step c: Create a DataFrame by joining the two Series by column
df = pd.DataFrame({'col1': series1, 'col2': series2})

# Display the DataFrame
print(df)

#12. Perform the following operations using people data set:

#a) Delete the 'Email', 'Phone', and 'Date of birth' columns from the dataset.

#b) Delete the rows containing any missing values.

#d) Print the final output also

import pandas as pd

# Sample data to mimic the 'people' dataset
data = {
    'Name': ['John Doe', 'Jane Doe', 'Alice Smith', 'Bob Brown'],
    'Email': ['john@example.com', 'jane@example.com', None, 'bob@example.com'],
    'Phone': ['123-456-7890', '987-654-3210', '555-555-5555', None],
    'Date of Birth': ['1990-01-01', '1992-02-02', '1994-03-03', '1995-04-04'],
    'Age': [31, 29, 27, 26]
}

# Create a DataFrame
people_df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(people_df)

# a) Delete the 'Email', 'Phone', and 'Date of Birth' columns
people_df.drop(columns=['Email', 'Phone', 'Date of Birth'], inplace=True)

# b) Delete the rows containing any missing values
people_df.dropna(inplace=True)

# d) Print the final output
print("\nFinal DataFrame after deletions:")
print(people_df)


# 13 .  Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:

#a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.

#b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.

#c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.

#d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.

#e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.

#f) Display a legend for the scatter plot, the horizontal line, and the vertical line.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1
np.random.seed(0)  # Optional: set seed for reproducibility
x = np.random.rand(100)
y = np.random.rand(100)

# Step 2: Create a scatter plot using x and y
plt.scatter(x, y, color='red', marker='o', label='Data Points')

# Step 3: Add a horizontal line at y = 0.5 using a dashed line style
plt.axhline(y=0.5, color='blue', linestyle='--', label='y = 0.5')

# Step 4: Add a vertical line at x = 0.5 using a dotted line style
plt.axvline(x=0.5, color='green', linestyle=':', label='x = 0.5')

# Step 5: Label the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Step 6: Set the title of the plot
plt.title('Advanced Scatter Plot of Random Values')

# Step 7: Display a legend for the scatter plot, the horizontal line, and the vertical line
plt.legend()

# Step 8: Show the plot
plt.show()

#14 . Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

#b) Label the x-axis as 'Date'.

#c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset
# Generate a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.normal(loc=20, scale=5, size=len(date_range))  # Mean 20, std deviation 5
humidity = np.random.normal(loc=60, scale=10, size=len(date_range))    # Mean 60, std deviation 10

# Create a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Plot Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(data['Date'], data['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Step 3: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Step 4: Show grid and plot
ax1.grid()
plt.show()

# 15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

#a) Plot a histogram of the data with 30 bins.

#b) Overlay a line plot representing the normal distribution's probability density function (PDF).

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

#d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
np.random.seed(0)  # Optional: for reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean 0, Standard deviation 1

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
# Define the x values for the PDF
xmin, xmax = plt.xlim()  # Get the limits from the histogram
x = np.linspace(xmin, xmax, 100)  # 100 points from xmin to xmax
p = norm.pdf(x, loc=0, scale=1)  # PDF of a normal distribution with mean 0 and std deviation 1

# Plot the PDF
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Step 6: Display the legend
plt.legend()

# Step 7: Show the plot
plt.show()


# 16. Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
np.random.seed(0)  # Optional: For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean 0, Standard deviation 1

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
# Define the x values for the PDF
xmin, xmax = plt.xlim()  # Get the limits from the histogram
x = np.linspace(xmin, xmax, 100)  # 100 points from xmin to xmax
p = norm.pdf(x, loc=0, scale=1)  # PDF of a normal distribution with mean 0 and std deviation 1

# Plot the PDF
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')  # Title set as requested

# Step 6: Display the legend
plt.legend()

# Step 7: Show the plot
plt.show()


# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create two random arrays of the same size
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 20 - 10  # Random values from -10 to 10
y = np.random.rand(100) * 20 - 10  # Random values from -10 to 10

# Step 2: Determine the quadrant for each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'
    else:
        return 'Origin'

# Create a DataFrame for easier plotting
data = pd.DataFrame({'x': x, 'y': y})
data['Quadrant'] = data.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Step 3: Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Step 4: Add a legend
plt.legend(title='Quadrants')

# Step 5: Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Step 6: Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Step 7: Show the plot
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Add y-axis line
plt.show()

# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function .

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Step 1: Prepare the data
x = np.linspace(0, 2 * np.pi, 100)  # X values from 0 to 2*pi
y = np.sin(x)  # Sine of X values

# Step 2: Create a new plot with a title and axis labels
output_notebook()  # Use this line if running in a Jupyter Notebook
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)',
           width=800, height=400)

# Step 3: Add a line renderer for the sine wave
p.line(x, y, legend_label='sin(x)', line_color='blue', line_width=2)

# Step 4: Add grid lines
p.grid.grid_line_color = 'gray'  # Color of grid lines
p.grid.grid_line_alpha = 0.5  # Transparency of grid lines

# Step 5: Show the plot
show(p)

# 19 . Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Step 1: Prepare the data
np.random.seed(0)  # For reproducibility
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))  # Random values for categories

# Step 2: Create the color mapping based on values
colors = [(0, 0, 255 * (value / max(values))) for value in values]  # Blue shades based on value
colors_hex = ['#%02x%02x%02x' % (int(r), int(g), int(b)) for r, g, b in colors]  # Convert RGB to HEX

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values, 'colors': colors_hex})

# Step 3: Create a ColumnDataSource
source = ColumnDataSource(data)

# Step 4: Create a new plot with a title and axis labels
output_notebook()  # Use if running in Jupyter Notebook
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

# Step 5: Add bars to the plot
bars = p.vbar(x='categories', top='values', width=0.9, source=source,
              line_color='white', fill_color='colors')

# Step 6: Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 7: Show the plot
show(p)

# 20. Usingg Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot

import plotly.graph_objects as go
import numpy as np

# Generate a random dataset
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # X values
y = np.random.random(100) * 100  # Random Y values scaled between 0 and 100

# Create a line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis Label',
    yaxis_title='Y Axis Label'
)

# Show the plot
fig.show()

# 21.  Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart

import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = np.random.randint(10, 100, size=len(labels))  # Random values for each category

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent', hole=0.3)])

# Update layout with title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()




```
